```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_homicidios= pd.read_excel('homicidios (1).xlsx')
```

```python
# Primero, convierte la columna de fecha a formato datetime si a√∫n no lo est√°
df_homicidios['FECHA'] = pd.to_datetime(df_homicidios['FECHA'])

# Filtrar el √∫ltimo semestre del a√±o 2021
ultimo_semestre_2021 = df_homicidios[(df_homicidios['FECHA'] >= '2021-07-01') & (df_homicidios['FECHA'] <= '2021-12-31')]
ultimo_semestre_2021
```

```python
# Calcular el n√∫mero de v√≠ctimas en el semestre anterior
homicidios_semestre_anterior = df_homicidios[(df_homicidios['FECHA'] >= '2021-01-01') & (df_homicidios['FECHA'] <= '2021-06-30')]
total_victimas_semestre_anterior = homicidios_semestre_anterior['N_VICTIMAS'].sum()

# Contar el n√∫mero de v√≠ctimas en el √∫ltimo semestre del a√±o 2021
total_victimas_ultimo_semestre_2021 = ultimo_semestre_2021['N_VICTIMAS'].sum()

print("El n√∫mero total de v√≠ctimas en el semestre anterior es:", total_victimas_semestre_anterior)
print("El n√∫mero total de v√≠ctimas en el √∫ltimo semestre del a√±o 2021 es:", total_victimas_ultimo_semestre_2021)
```

```python
# Filtrar los accidentes donde la columna 'victima' indica que est√° involucrada los autos
accidentes_auto = df_homicidios[df_homicidios['VICTIMA'].str.contains('auto', case=False, na=False)]

# Filtrar los accidentes mortales (donde hay al menos una v√≠ctima)
accidentes_mortales_auto = accidentes_auto[accidentes_auto['N_VICTIMAS'] > 0]

# Contar la cantidad de accidentes mortales de motociclistas
cantidad_accidentes_mortales_moto = len(accidentes_mortales_auto)

print("La cantidad de accidentes mortales de auto es:", cantidad_accidentes_mortales_moto)
```

--> La cantidad de accidentes mortales en auto son 83

```python
# 1. Filtrar los accidentes que involucren motocicletas
accidentes_auto = df_homicidios[df_homicidios['VICTIMA'].str.contains('auto', case=False, na=False)]

# 2. Crear una copia expl√≠cita del DataFrame antes de modificarlo
accidentes_auto = accidentes_auto.copy()

# 3. Dividir los datos por a√±o
# Suponiendo que tienes una columna 'FECHA' en formato datetime
accidentes_auto['AAAA'] = accidentes_auto['FECHA'].dt.year

# 4. Contar el n√∫mero de accidentes mortales con v√≠ctimas en moto para cada a√±o
accidentes_por_a√±o = accidentes_auto.groupby('AAAA').agg({'N_VICTIMAS': 'sum'})
accidentes_por_a√±o.columns = ['accidentes_mortales_auto']
accidentes_por_a√±o.reset_index(inplace=True)

# 5. Obtener el n√∫mero de accidentes mortales con v√≠ctimas en moto para el a√±o anterior y el a√±o actual
accidentes_anio_actual = accidentes_por_a√±o[accidentes_por_a√±o['AAAA'] == 2021]['accidentes_mortales_auto'].values[0]
accidentes_anio_anterior = accidentes_por_a√±o[accidentes_por_a√±o['AAAA'] == 2020]['accidentes_mortales_auto'].values[0]

print("N√∫mero de accidentes mortales con v√≠ctimas en moto en el a√±o anterior:", accidentes_anio_anterior)
print("N√∫mero de accidentes mortales con v√≠ctimas en moto en el a√±o actual:", accidentes_anio_actual)
```

--> N√∫mero de accidentes mortales con v√≠ctimas en moto en el a√±o anterior: 12

--> N√∫mero de accidentes mortales con v√≠ctimas en moto en el a√±o actual: 11

```python
(accidentes_anio_anterior - accidentes_anio_actual) - (accidentes_anio_anterior) * 100
```

```python
# Calcular la cantidad de accidentes mortales de motociclistas que se necesitan para alcanzar una reducci√≥n del 7%
reduccion_objetivo = 7
reduccion_actual = ((accidentes_anio_anterior - accidentes_anio_actual) / accidentes_anio_anterior) * 100
cantidad_a_reducir = ((reduccion_objetivo / 100) * accidentes_anio_anterior)

# Calcular el n√∫mero objetivo de accidentes mortales de motociclistas para cumplir con la reducci√≥n del 7%
numero_objetivo = accidentes_anio_actual + cantidad_a_reducir

print("Para reducir en un 7% la cantidad de accidentes mortales de motociclistas en el √∫ltimo a√±o, se necesitan reducir", cantidad_a_reducir, "accidentes.")
print("El n√∫mero objetivo de accidentes mortales en auto para cumplir con la reducci√≥n del 7% es:", numero_objetivo)
```

--> Para reducir en un 7% la cantidad de accidentes mortales en auto en el √∫ltimo a√±o, se necesitan reducir 11.84 accidentes.
